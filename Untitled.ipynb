```python
import numpy as np
import cv2
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import glob

%matplotlib inline
```

```python
gestures = ['a', 'b', 'c', 'd', 'g', 'h', 'i', 'l', 'v', 'y']
X_all = []

y_all = []
paths = []

prefix = 'Triesch/*'
suffix = '.pgm'

for char in gestures:
    paths_char = []
    for i in range(1,4):
        paths_char.extend(glob.glob(prefix + char + str(i) + suffix))
    y_all.extend(([char] * len(paths_char)))
    paths.extend(paths_char)
    
print (paths)
print (y_all)
```

```python
for path in paths:
    X_all.append(cv2.equalizeHist(mpimg.imread(path)) * 1./255 - 0.5)

print (len(X_all))
print (len(y_all))
```

```python
X_all = np.array(X_all)
y_all = np.array(y_all)

print (y_all[:5])
from sklearn.preprocessing import OneHotEncoder, LabelEncoder, LabelBinarizer

enc = LabelEncoder()

y_all = enc.fit_transform(y_all)

print (y_all[:5])
```

```python
from keras.models import Sequential
import keras.layers
```

```python
model = Sequential()

model.add(keras.layers.Reshape(input_shape=(128, 128), target_shape=(128,128,1)))
model.add(keras.layers.Conv2D(input_shape=(128,128,1), filters=32, kernel_size=3, strides=1, activation='tanh', kernel_initializer='glorot_uniform'))
model.add(keras.layers.Conv2D(filters=32, kernel_size=3, strides=1, padding='same', activation="tanh", kernel_initializer='glorot_uniform'))
model.add(keras.layers.MaxPooling2D())

model.add(keras.layers.Conv2D(filters=64, kernel_size=3, strides=1, padding='same', activation="tanh", kernel_initializer='glorot_uniform'))
model.add(keras.layers.Conv2D(filters=64, kernel_size=3, strides=1, padding='same', activation="tanh", kernel_initializer='glorot_uniform'))
model.add(keras.layers.MaxPooling2D())

model.add(keras.layers.Conv2D(filters=128, kernel_size=3, strides=1, padding='same', activation="tanh", kernel_initializer='glorot_uniform'))
model.add(keras.layers.Conv2D(filters=128, kernel_size=3, strides=1, padding='same', activation="tanh", kernel_initializer='glorot_uniform'))
model.add(keras.layers.MaxPooling2D())

model.add(keras.layers.Flatten())
model.add(keras.layers.Dense(16, activation='relu'))
model.add(keras.layers.Dense(10, activation='softmax'))

model.summary()
```

```python
from keras.optimizers import SGD

sgd = SGD(lr=0.01, decay=1e-6, momentum=0.9, nesterov=True)
model.compile(loss='categorical_crossentropy', optimizer=sgd, metrics=['accuracy'])
```

```python
tensorb = keras.callbacks.TensorBoard(log_dir='./logs', histogram_freq=0, batch_size=None, write_graph=True, write_grads=False, write_images=False)

history = model.fit(x=X_train, y=y_train, batch_size=None, epochs=15, verbose=1, validation_data=(X_val, y_val), shuffle=True, callbacks=[tensorb])
```

```python
model.fit(x=X_all, y=y, batch_size=None, epochs=20, verbose=1, validation_split=0.15, shuffle=True, callbacks=[tensorb], initial_epoch=15)
model.save('shape-128-equalized-june11.h5')
```
